```python
# unique() works only for series
joined[['Brand','Pack-Type','Pack-Size','SKU']].drop_duplicates()
```

```python
vol_disc_add= joined.groupby(['Year', 'Month','Brewery', 'Brand', 'Segment', 'Pack-Type',
       'Pack-Size', 'SKU'],sort=False)['Volume','Activity', 'Invoice', 'Target Bonus'].sum().reset_index()
```

```python
vol_disc_add
```

```python
temp=joined.groupby(['Year', 'Month','Brewery', 'Brand', 'Segment', 'Pack-Type',
       'Pack-Size', 'SKU'],sort=False)['Dealer Price', 'Dealer-Commission',
       'VAT', 'W/s-Price'].nunique().reset_index()
```

```python
temp
```

```python
temp[temp['Dealer Price']>1]
# If a boolean is given, only that is selected

temp[temp['Dealer-Commission']>1].shape # 0 rows and 12 columns. 

(temp['VAT']>1).sum() # adds all trues and false. True= 1 and false = 0

(temp['W/s-Price']>1).sum() 
# Hence we can confirm we only have unique values in the columns.

# Shift+M means merging dfferent parts of the code 
```

```python
price_same=joined.groupby(['Year', 'Month','Brewery', 'Brand', 'Segment', 'Pack-Type',
       'Pack-Size', 'SKU'],sort=False)['Dealer Price', 'Dealer-Commission',
       'VAT', 'W/s-Price'].mean().reset_index()
# first, last, mean, max, min----- Same output in this case. All data is same

```

```python
price_same
```

```python
VPD_final=pd.merge(vol_disc_add,price_same,how='left',on=['Year', 'Month','Brewery', 'Brand', 'Segment', 'Pack-Type',
       'Pack-Size', 'SKU'])
```

```python
VPD_final
```

```python
mark=pd.read_csv(r'E:\Documents_internship\Carlsberg\Data\Data_Extra\Modified\From Sharepoint\Marketing_Cider_Removed.csv')
# mark
```

```python
mark_pivot=mark.pivot(columns='Discount Type', values='Amount')
# mark_pivot
```

```python
mark=pd.concat([mark,mark_pivot],axis=1)
mark
```

```python
mark.drop(['Discount Type','Amount'],axis=1,inplace=True)
```

```python
mark.rename(columns = {'Activity':'Activity_Marketing','SKU':'Brand'},inplace=True) 
# Always use underscore
#Avoid whitespaces
```

```python
mark_temp=mark.groupby(['Year', 'Month','Brand'],sort=False)['Activity_Marketing',
       'Brand Marketing', 'Trade Marketing'].sum().reset_index()
```

```python
VPDM=pd.merge(VPD_final,mark_temp, how= 'left', on=['Year', 'Month','Brand'])
```
